```python
import pandas as pd

df = pd.DataFrame({
    'price': [20000, 30000, 40000],
    'discount': [2000, 3000, 5000]
})

# New column: final price
df['final_price'] = df['price'] - df['discount']
print(df)
```

CREATING PRICE CATEGORY

```python
import numpy as np

df['price_category'] = np.where(df['price'] > 25000, 'Expensive', 'Affordable')
print(df)
```

CREATING RATING

```python
def classify_price(price):
    if price >= 35000:
        return 'High'
    elif price >= 25000:
        return 'Medium'
    else:
        return 'Low'

df['price_level'] = df['price'].apply(classify_price)
print(df)
```

This process helps in validating assumptions, selecting relevant features, and making informed decisions about data cleaning and preprocessing. GROUPBY ANALYSIS

Groupby is a fundamental technique for performing "split-apply-combine" analysis. It allows you to partition a DataFrame into smaller segments based on specific categories, perform a calculation on those segments, and then merge the results back into a summary table. This is essential for comparing different subsets of data, such as finding the average salary by department or the total revenue per product category.

PIVOT TABLE

Pivot Table is a powerful tool used to summarize, sort, and reorganize large datasets. It works by transforming "long" data into a "wide" format, allowing you to group data by one or more keys and apply an aggregation functionlike a sum or averageto the values. This makes it much easier to spot trends and patterns that are hidden in a flat table.

```python
pd.pivot_table(
    df,
    values='price',
    index='battery',
    aggfunc='mean'
)
```

Correlation Analysis is a statistical technique used to evaluate the strength and direction of the relationship between two or more continuous variables. By using the .corr() method on a DataFrame, pandas calculates a correlation coefficientmost commonly the Pearson coefficientfor every pair of numerical columns. This resulting "Correlation Matrix" is a vital tool during the Exploratory Data Analysis

HISTOGRAM

A Histogram is a graphical representation that organizes a group of data points into user-specified ranges. It is the most common tool used to visualize the frequency distribution of a continuous numerical datase

```python
plt.figure(figsize=(8,5))
plt.hist(df["price"], bins=20)
plt.title("Distribution of Smartphone Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

SCATTER PLOT - RAM VS PRICE

A Scatter Plot is a 2D data visualization that uses dots to represent the values obtained for two different variablesone plotted along the x-axis and the other along the y-axis. In pandas and data science, scatter plots are the primary tool for investigating the relationship or correlation between numerical features.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="ram", y="price", data=df)
plt.title("RAM vs Price")
plt.xlabel("RAM (GB)")
plt.ylabel("Price")
plt.show()
```

BOX PLOT- PRICE DISTRIBUTION

A Box Plot (or box-and-whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary: the minimum, first quartile (Q_1), median, third quartile (Q_3), and maxim

```python
plt.figure(figsize=(6,5))
sns.boxplot(y=df["price"])
plt.title("Box Plot of Smartphone Prices")
plt.ylabel("Price")
plt.show()
```

LINE CHART

A Line Chart (or line graph) is a type of data visualization that displays information as a series of data points called 'markers' connected by straight line segments. It is the most common tool used to visualize how a specific variable changes over a continuous interval, typically time.

```python
df_sorted = df.sort_values("battery")

plt.figure(figsize=(8,5))
plt.plot(df_sorted["battery"], df_sorted["price"])
plt.title("Price Trend by Battery")
plt.xlabel("Battery (mAh)")
plt.ylabel("Price")
plt.show()
```

RAM vs Price Relationship
